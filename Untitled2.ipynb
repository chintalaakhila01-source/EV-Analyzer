```python
import random
from datetime import datetime, timedelta
import pandas as pd

rows = []

stations = ["ST001", "ST002", "ST003", "ST004", "ST005"]
vehicles = [f"V{str(i).zfill(3)}" for i in range(1, 201)]

start_date = datetime(2025, 4, 1)

for i in range(1, 2001):
    station = random.choice(stations)
    vehicle = random.choice(vehicles)

    start_time = start_date + timedelta(
        days=random.randint(0, 9),
        hours=random.randint(0, 23),
        minutes=random.randint(0, 59)
    )

    duration = random.randint(30, 120)  # minutes
    end_time = start_time + timedelta(minutes=duration)

    energy = round(random.uniform(10, 25), 2)
    cost = round(energy * 15, 2)

    rows.append([
        i, station, vehicle,
        start_time, end_time,
        energy, cost
    ])

df = pd.DataFrame(rows, columns=[
    "session_id", "station_id", "vehicle_id",
    "start_time", "end_time",
    "energy_kwh", "cost"
])

df.head()
df.to_csv("ev_charging_2000.csv", index=False)
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("EV_Project").getOrCreate()
```

```python
spark_df = spark.createDataFrame(rows, [
    "session_id","station_id","vehicle_id",
    "start_time","end_time","energy_kwh","cost"
])

spark_df.show()
```

```python
from pyspark.sql.functions import hour, count, avg

# Peak hours
spark_df.groupBy(hour("start_time")).agg(count("*")).show()

# Avg energy
spark_df.groupBy("station_id").agg(avg("energy_kwh")).show()
```

```python
import matplotlib.pyplot as plt
pdf = spark_df.toPandas()
pdf.groupby("station_id")["energy_kwh"].mean().plot(kind='bar')
plt.title("Average Energy Consumption per Station")
plt.xlabel("Station ID")
plt.ylabel("Energy (kWh)")
plt.show()
```

```python
pdf["hour"] = pdf["start_time"].dt.hour

pdf.groupby("hour").size().plot(kind='line')
plt.title("Charging Sessions by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Sessions")
plt.show()
```

```python
pdf["energy_kwh"].plot(kind='hist', bins=10)
plt.title("Energy Consumption Distribution")
plt.xlabel("Energy (kWh)")
plt.show()
```

```python
pdf.groupby("station_id")["energy_kwh"].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title("Energy Share by Station")
plt.ylabel("")
plt.show()
```
